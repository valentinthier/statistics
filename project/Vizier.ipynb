```python
from astroquery.vizier import Vizier
import astropy.units as u
from astropy.table import Table

# Catalog ID from Vizier
CATALOGUE = "J/A+A/544/A53"

# "*" means fetch all columns
vizier = Vizier(columns=["*"], row_limit=-1)

# Get all tables from the catalog
result = vizier.get_catalogs(CATALOGUE)

# result is a list of tables
print(f"Number of tables in catalog: {len(result)}")

table = result[0]  # first table
print(table)       # prints all rows (might be long)
print(table.colnames)  # list all column names

df = table.to_pandas()
print(df)
df.describe()


table.write("vizier_data.csv", format="csv", overwrite=True)
```
