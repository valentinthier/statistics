```python
from astroquery.vizier import Vizier
import os

os.makedirs("Data", exist_ok=True)

CATALOGUE = "J/A+A/544/A53"
vizier = Vizier(columns=["*"], row_limit=-1)
result = vizier.get_catalogs(CATALOGUE)


m = result[0]                     
s = result[1]
p = result[2]
d_p1 = result[3]
d_p3 = result[4]
n = result[5]


magnitudes = m.to_pandas()
magnitudes.to_csv("Data/Magnitudes", index=False)

spectra = s.to_pandas()
spectra.to_csv("Data/Spectra", index=False)

photometry = p.to_pandas()
photometry.to_csv("Data/BV(RI)c photometry", index=False)

dphotometry1 = d_p1.to_pandas()
dphotometry1.to_csv("Data/Differential photometry (1 comparison star)", index=False)

dphotometry3 = d_p3.to_pandas()
dphotometry3.to_csv("Data/Differential photometry (3 comparison stars)", index=False)

info = n.to_pandas()
info.to_csv("Data/Observation informations", index=False)


'''
How to access later:

import panda as pd

some name = pd.read.csv("data/some file name.csv")

done

and then you can print for example

print(some name)
'''
```
